```python
import pandas as pd
import glob
import os
```

```python
df = pd.read_csv('G:/My Drive/GitHubProjects/MLS/data/scraping/matches/feed/feed_nshvscin-03-29-2025.csv')
```

```python
## remove row is minute+title+comment is nan

df = df.dropna(subset=['minute', 'title', 'comment'], how='all')
```

```python
## if corner in comment, then title is corner

df['title'] = df.apply(lambda x: 'Corner' if pd.notna(x['comment']) and 'corner' in x['comment'].lower() else x['title'], axis=1)

df['title'] = df.apply(lambda x: 'Foul' if pd.notna(x['comment']) and 'foul' in x['comment'].lower() else x['title'], axis=1)

df['title'] = df.apply(lambda x: 'Offside' if pd.notna(x['comment']) and 'offside' in x['comment'].lower() else x['title'], axis=1)
```

```python
### delete row is 'lineups' is in comment

df = df[~df['comment'].str.contains('Lineups', na=False)]

df = df[~df['title'].str.contains('KICK OFF|HALF TIME|FULL TIME|END OF SECOND HALF', na=False)]
```

```python
## if minute is missing, delete

df = df[df['minute'].notna()]
```

```python
### reverse minute sorting

df = df.iloc[::-1].reset_index(drop=True)
```

```python
def clean_feed(df):
    df = df.copy()

    df = df.dropna(subset=['minute', 'title', 'comment'], how='all')
    
    df['title'] = df.apply(lambda x: 'Corner' if pd.notna(x['comment']) and 'corner' in x['comment'].lower() else x['title'], axis=1)

    df['title'] = df.apply(lambda x: 'Foul' if pd.notna(x['comment']) and 'foul' in x['comment'].lower() else x['title'], axis=1)

    df['title'] = df.apply(lambda x: 'Offside' if pd.notna(x['comment']) and 'offside' in x['comment'].lower() else x['title'], axis=1)   
    
    df = df[~df['comment'].str.contains('Lineups', na=False)]

    df = df[~df['title'].str.contains('KICK OFF|HALF TIME|FULL TIME|END OF SECOND HALF', na=False)]
    
    df = df[df['minute'].notna()]
    
    df = df.iloc[::-1].reset_index(drop=True)
    
    return df
```

```python
files = glob.glob('data/scraping/matches/feed/*.csv')
```

```python
for file in glob.glob('G:/My Drive/GitHubProjects/MLS/data/scraping/matches/feed/*.csv'):
    df = pd.read_csv(file)
    print(f'Cleaning {file}...')
    cleaned_df = clean_feed(df)
    filename = os.path.basename(file)
    cleaned_df.to_csv(
        f"G:/My Drive/GitHubProjects/MLS/data/data_clean/matches/feed/cleaned_{filename}.csv",
        index=False)
    print(f"Cleaned {file} and saved the result.")
```
