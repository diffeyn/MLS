```python
## rename columns
df = df.rename(columns={
    'Player': 'player_name',
    'Mins': 'minutes',
    'G' : 'goals',
    'xG': 'expected_goals',
    'Conv%' : 'shot_conv_perc',
    'SOT' : 'on_target',
    'Pass%' : 'pass_perc',
    'A' : 'assists',
    'P' : 'passes',
    'Cross' : 'cross',
    'CK' : 'corner_kick',
    'KP' : 'key_pass',
    'AD' : 'aerial',
    'AD%' : 'aerial_perc',
    'FC' : 'fouls',
    'FS' : 'fouls_against',
    'OFF' : 'offside',
    'YC' : 'yellow_card',
    'RC' : 'red_card'
})
```

```python
## move match_id to the front

df = df[['match_id'] + [col for col in df.columns if col != 'match_id']]
```

```python
def clean_players(df):
    df = df.rename(columns={
        'Player': 'player_name',
        'Mins': 'minutes',
        'G' : 'goals',
        'xG': 'expected_goals',
        'Conv%' : 'shot_conv_perc',
        'SOT' : 'on_target',
        'Pass%' : 'pass_perc',
        'A' : 'assists',
        'P' : 'passes',
        'Cross' : 'cross',
        'CK' : 'corner_kick',
        'KP' : 'key_pass',
        'AD' : 'aerial',
        'AD%' : 'aerial_perc',
        'FC' : 'fouls',
        'FS' : 'fouls_against',
        'OFF' : 'offside',
        'YC' : 'yellow_card',
        'RC' : 'red_card'
    })
    
    df = df[['match_id'] + [col for col in df.columns if col != 'match_id']]
    
    return df
```

```python
for file in glob.glob('G:/My Drive/GitHubProjects/MLS/data/scraping/matches/players/*.csv'):
    df = pd.read_csv(file)
    cleaned_df = clean_players(df)
    filename = os.path.basename(file)
    cleaned_df.to_csv(
        f"G:/My Drive/GitHubProjects/MLS/data/data_clean/matches/players/cleaned_{filename}.csv",
        index=False
    )
    
    
```
