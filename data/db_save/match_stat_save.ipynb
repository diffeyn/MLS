```python
def rem_teams(df):
    if 'teams_home' in df.columns:
        df.drop(columns=['teams_home', 'teams_away', 'match_date'], inplace=True)
    cols = df.columns.tolist()
    cols.insert(0, cols.pop(cols.index('match_id_hash')))
    df = df[cols]
    df.rename(columns={'match_id_hash': 'match_id'}, inplace=True)
    return df

df2 = rem_teams(df2)

df2
```

```python
folder = glob.glob('G:/My Drive/GitHubProjects/MLS/data/data_clean/matches/w_match_id/stats/*.csv')
matches = []
for file in folder:
    if file.endswith('.csv'):
        df = pd.read_csv(file)
        df = rem_teams(df)
        matches.append(df)

df = pd.concat(matches, ignore_index=True)
df
```

```python
## print full list of df colnames downwards for easier reading

for col in df.columns.tolist():
    print(col)
```

```python
### check columns if they are actually float by checking values after the decimal point

for col in df.select_dtypes(include=['float']).columns:
    if df[col].astype(str).str.contains('.').any():
        if (df[col] % 1 != 0).any():
            print(f"Column '{col}' is float with decimal values")
        else:
            print(f"Column '{col}' is float without decimal values")
            df[col] = df[col].astype(int)


```

```python
## concat clean match data into one dataframe

input_folder = 'G:/My Drive/GitHubProjects/MLS/data/data_clean/matches/w_match_id/simple_stats'

all_files = [f"{input_folder}/{i}" for i in os.listdir(input_folder) if i.endswith('.csv')]

matches = []
for file in all_files:
    df = pd.read_csv(file)
    matches.append(df)
    
matches = pd.concat(matches, ignore_index=True)
```
